```python
import pandas as pd
from unidecode import unidecode
```

## Detalle de puntos de recarga, ubicación geográfica

```python
recargas = pd.read_csv("https://servicios.transporte.gob.ar/gobierno_abierto/descargar.php?t=red&d=recarga", sep=";", encoding="UTF-8")
recargas
```

```python
#NOBRE DE COLUMNAS
recargas.columns = recargas.columns.str.upper()
```

```python
recargas.columns
```

```python
recargas[['DIRECCION', 'LOCALIDAD', 'PROVINCIA', 'TIPO_UBICACION']] = recargas[['DIRECCION', 'LOCALIDAD', 'PROVINCIA', 'TIPO_UBICACION']].applymap(lambda x: unidecode(x).upper())
```

```python
recargas.to_csv("./data/03-TRENES-PUNTOS-RECARGAS.csv", index_label="INDEX")
```
